# 멱집합

- 주어진 집합에서 가능한 모든 부분 집합들의 집합
- example
    - data={a, b, c, d}
    - powerset: {공}, {a}, {b}, {c}, {d}, {a,b}, {a,c}, {a,d}
    - , {b,c}, {b,d}, {c,d}, {a,b,c}, {a,b,d}, {a,c,d}, {b,c,d}, {a,b,c,d}

#### {a, b, c, d, e, f}의 모든 powerset을 나열하려면....

- a를 제외한 {b, c, d, e, f}의 모든 부분집합을 나열하고,
- {b, c, d, e, f}의 모든 부분집합에 {a}를 추가한 집합들을 나열한다.

{a, b, c, d, e, f} = $2^5$(a를 포함안한 subset) * $2^5$(a를 포함한 subset) = $2^6$

### Pseudo code


- S의 멱집합을 출력
- if 집합 S가 공집합이면 아무것도 출력 안함.
- else 집합 S의 첫번째 원소를 t라고 하자
- 집합 S에서 t를 뺀 모든 집합에 대한 powerset함수 recursion하여 찾는다.
- 그 subset을 출력한다.
- 그 후, t를 더하여 subset을 출력한다.

==> 위의 방법은 모든 subset을 찾은 후 return하여 출력하기 때문에 매우 비효율

#### 해결안 Resolution


- S의 멱집합을 구한 다음 더해줄 P라는 집합도 매개변수로 넣어준다.
- if:    집합 S가 공집합이면, 집합 P를 출력한다.
- else:  집합 S의 첫 원소를 t라 한다
- 멱집합에 합쳐줄 P집합과 집합 S에서 원소 t를 하나뺀 집합에 recursion
- 그 후 원소 t 와 집합 P를 합집합한 집합과 t를 하나뺀 집합에 recursion


- 결국, 집합 P를 출력하는 것이며,
- 이 집합 P는 처음에 공집합을 출력하고, 점차 t가 채워지면서 계속 출력해 나갈것이다.

#### 표현


### CODE

```python
data = ['a', 'b', 'c', 'd', 'e', 'f']
data2 = ['a', 'b', 'c']

n = len(data)
n2 = len(data2)

include = [0, 0, 0, 0, 0, 0]
include2 = [0, 0, 0]
```

```python
def powerSet(k):    # 내가 노드의 어느 레벨에 있는가?
    if (k==n):      # n = 6 만약 내가 리프노드라면...
        for i in range(n):
            if include[i]:
                print(data[i], end=" ")
        print("\n")
        return
    include[k] = False
    powerSet(k+1)    
    include[k] = True
    powerSet(k+1)    
    
    
    

def powerSet2(k):
    if (k==n2):    # n = 3
        for i in range(n2):
            if include2[i]:
                print(data2[i], end=" ")
        print("\n")
        return
    include2[k] = False
    powerSet2(k+1)
    include2[k] = True
    powerSet2(k+1)
```

```python
#powerSet(0)
print('0')
powerSet2(0)
```
