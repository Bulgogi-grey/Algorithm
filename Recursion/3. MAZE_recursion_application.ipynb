# MAZE (미로찾기)

- 맵 표현


- 입구로 들어와 출구까지 나가야합니다.
- Wall에는 들어갈 수 없으며, pathway만 지나가게됩니다.

#### Recursive thinking
- 현재 위치에서 출구까지 가는 경로가 있으려면...
    - 현재 위치가 출구이거나 
    - 이웃한 셀들 중 하나에서 현재 위치를 지나지 않고 출구까지 가는 경로가 있거나..

- 문제를 해결하기 위해 Decision problem을 정의해야한다.  
** decision problem은 답이 yes 이거나 no인 문제


- findPath(x, y)에 대한 decision problem


- x, y를 매개변수로 받는다. x, y는 나의 위치.
- if 내가 있는 곳이 출구라면, true를 반환... true는 탈출했다는 의미
- else x,y에서 이웃한 x',y' 에 대하여...
        - 만약 x',y'이 갈 수 있는 길이면
        - 건너간 후, 
        - if x', y'가 출구이면,
            - true
        - 건나가서 봤는데 출구가 아니면,
            - false

#### CODE

```python
import numpy as np

def find_maze_path(x, y):
    if x<0 or y<0 or x>= N or y>= N:
        return False
    elif maze[x, y] != PATHWAY:
        return False
    elif x == N-1 and y == N-1:
        maze[x][y] = PATH
        return True
    else:
        maze[x][y] = PATH
        if find_maze_path(x-1, y) or find_maze_path(x, y+1) or find_maze_path(x+1, y) or find_maze_path(x, y-1):
            return True
        maze[x, y] = BLOCKED
        return False
    
maze = np.zeros([8, 8])

PATHWAY = 0
WALL = 1
BLOCKED = 2
PATH = 3
N = 8
```

```python
maze[:, -1] = WALL
maze[[3,-1], -1] = 0
maze[1, [1, 2, 4, 5]] = WALL
maze[2,3] = WALL
maze[3, [1, 4, 5]] = WALL
maze[4, [1, 2, 3, 6]] = WALL
maze[5, [1, 5]] = WALL
maze[6, 3] = WALL
maze[7, [1, 2, 3, 5]] = WALL
```

```python
print(maze)
print(find_maze_path(0, 4))
print(maze)
```
