# 몇명의 여왕을 판안에 둘 수 있을까?

- 체스판에 여왕들을 안겹치게 둘 수 있는 방법을 recursion으로 풀어내자!

#### Backtracking
- 내가 해온 선택을 번복하고 다시 하는 방법

- Backtracking을 시각화하면,,, 상태공간트리라는 개념이 나온다


#### Backtracking을 구현하는 방법
    - recursion
    - stack

- recursion pseudo Code

- Queens을 놓을 수 있는 위치를 찾는 함수
    - if 그 자리에 놓을 수 없다면
        - 실패하고 return
    - elif 놓을 수 있다면
        - 성공하고 return
    - else(둘 다 아니면==일반적인 경우) 다른 child node로 recursion

### CODE

```python
import numpy as np

def promising(level):
    for i in range(1, level):
        if cols[i] == cols[level]:
            return False
        elif (level - i) == abs(cols[level]  - cols[i]):
            return False
        
    return True
    
def queens(level):
    if not promising(level):
        return False
    elif level == N:
        print(cols)
        return True
    
    for i in range(1, N+1):
        cols[level+1] = i
        if (queens(level+1)):
            return True
    return False

N=4
cols = [0, 0, 0, 0, 0]
queens(0)
```
