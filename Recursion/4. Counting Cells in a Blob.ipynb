# 그림 안에서 연결된 셀 찾기

- 영상처리에서의 4연결법,,, 8연결법과 같은 알고리즘
- 이 알고리즘을 영상처리의 4연결법은 marking을 하면서 그림을 한 필셀씩 움직이면서 몇개의 셀이 있는지 확인한다.

- Objective: 픽셀의 위치가 주어지면, 그 픽셀에 cell이 있는지 확인하고 있다면 그 셀의 크기를 반환.

```python
def find_cells_scale(x, y):
    
    if x<0 or y<0 or x>=N or y>=N:
        return 0
    elif img[x, y] != IMAGE:
        return 0
    else:
        img[x, y] = ALREADY
        return 1+ find_cells_scale(x-1, y+1)+ find_cells_scale(x, y+1) + find_cells_scale(x+1, y+1) + find_cells_scale(x-1, y) + find_cells_scale(x+1, y) + find_cells_scale(x-1, y-1) + find_cells_scale(x, y-1) + find_cells_scale(x+1, y-1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = np.array([[50, 0, 0, 0, 0, 0, 0, 50],
                [0, 50, 50, 0, 0, 50, 0, 0],
                [50, 50, 0, 0, 50, 0, 50, 0],
                [0, 0, 0, 0, 0, 50, 0, 0],
                [0, 50, 0, 50, 0, 50, 0, 0],
                [0, 50, 0, 50 ,0, 50 ,0 ,0],
                [50, 0, 0, 0, 50,0 ,0 , 50],
                [0, 50, 50, 0, 0, 50, 50,50]]
                )

IMAGE = 50
ALREADY = 2
N =8
```

```python
plt.imshow(img)
result = find_cells_scale(0, 0)
print(f"the scale of cell: {result}")
```
