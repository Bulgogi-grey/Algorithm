# recursion에 대한 구현

- recursion: 자기 자신을 호출하는 함수

```python
import numpy as np

def func():
    print('Hi')
    func()
    
if __name__ == '__main__':
    func()
```

- 위 함수는 recursion의 예시로, 무한 루프에 빠지게 된다. 함수가 끝나지 않는다는 것.

### recursion은 항상 무한루프에 빠질까?

- 아니다, 함수에 쓰는 내용에 따라 달라진다.

```python
def func(k):
    if k <= 0:
        return
    else:
        print('hi')
        func(k-1)

if __name__=='__main__':
    func(4)
```

- 따라서 무한루프에 빠지지 않으려면 다음과 같은 조건을 가져야한다.

- 조건1: 적어도 하나의 자기자신을 호출하지 않는 경우가 존재해야하며, 이를 base case라고 한다.
- 조건2: recursion case가 결국 나중에 base case로 수렴해야만 한다.

### recursion의 다양한 예시

#### n~n-1까지의 합

```python
def sum_(n):
    if n==0:
        return 0
    else:
        return n+sum_(n-1)

if __name__ == '__main__':
    print(sum_(4))    
```

- 위의 함수는 수학적 귀납법과 일치한다.
- 따라서 수학적 증명과 해석이 가능하다.

- 다음은 해당 증명이다.


- 정리: sum_(n)은 음이 아닌 정수 n에 대해서 n까지의 합을 올바로 계산한다.
    - n = 0인 경우: n=0인 경우 0을 반환한다.
    - 임의의 양의 정수 k에 대해서 n<k인 경우 0에서 n까지의 합을 올바르게 계산한다고 가정
    - n=k인 경우: sum_은 먼저 sum_(n-1)을 호출하는데 2번의 가정에 의해서 0에서 n-1까지의 합이 계산되어 반환된다.
    - 따라서 sum_(n)은 n + (0~n-1 까지의 합)을 반환한다.

#### Factorial

```python
def f(n):
    if n==0:
        return 1
    else:
        return n*f(n-1)
    
if __name__ == '__main__':
    print(f(4))
```

#### X의 n승

```python
def x_square(x, n):
    if n == 0:
        return 1
    else:
        return x*x_square(x, n-1)
    
if __name__=='__main__':
    print(x_square(2, 4))
```

#### Fibonacci

```python
def f(n):
    """if n == 0:
        return 0
    elif n == 1:
        return 1"""
    if n < 2:
        return n
    else:
        return f(n-1)+f(n-2)
        
if __name__ == '__main__':
    print(f(6))
```

#### Euclid Method (최대공약수)

```python
def gcd(m, n):    # gcd: greatest common divider
    low, high = min(m, n), max(m, n)
    
    if  high % low == 0:
        return low
    else:
        return gcd(low, high%low)

if __name__=='__main__':
    
    result = gcd(41,10)
    
    if result ==1:
        print('It is coprime')
    else:
        print(result)
```
